```python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
    api_key=os.environ['api_key'],  # 如果您没有配置环境变量，请在此处用您的API Key进行替换
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1"  # 百炼服务的base_url
)
```

```python
def get_embedding(text, model="text-embedding-v3"):
   text = text.replace("\n", " ")
   return client.embeddings.create(input = [text], model=model).data[0].embedding
```

```python
import numpy as np

def cosine_similarity(A, B):
    dot_product = np.dot(A, B)
    norm_A = np.linalg.norm(A)
    norm_B = np.linalg.norm(B)
    return dot_product / (norm_A * norm_B)
```

```python
emb1 = get_embedding("大模型的应用场景很多")
emb2 = get_embedding("python的应用场景很多")
emb3 = get_embedding("大模型有很多应用场景")
emb4 = get_embedding("python开发")
```
