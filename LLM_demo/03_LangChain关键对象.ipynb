# 输出格式化

```python
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.prompts import PromptTemplate
from pydantic import BaseModel, Field

# 定义数据结构
class PersonInfo(BaseModel):
    name: str = Field(description="人的姓名")
    age: int = Field(description="人的年龄")

# 创建解析器
parser = PydanticOutputParser(pydantic_object=PersonInfo)

# 创建提示模板
template = """你是一个信息提取助手，请从以下文本中提取姓名和年龄。
文本:{input_text}
{format_instructions}"""

prompt = PromptTemplate(
    template=template,
    partial_variables={"format_instructions": parser.get_format_instructions()}
)

# 使用
input_text = "张三今年25岁，来自北京。"
filled_prompt = prompt.format(input_text=input_text)

# 调用模型
response = chat.invoke(filled_prompt)

# 解析（返回 PersonInfo 对象）
result = parser.parse(response.content)
print(result.name, result.age)  # 张三 25
```

# 链式调用

```python
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.prompts import ChatPromptTemplate
from pydantic import BaseModel, Field

# 定义输出结构（使用 Pydantic）
class MathAnswer(BaseModel):
    answer: str = Field(description="问题的答案")
    confidence: float = Field(description="答案的置信度，范围0-1")

# 创建输出解析器
output_parser = PydanticOutputParser(pydantic_object=MathAnswer)

# 定义提示模板
template = """你是一名数学老师，请用{style}风格回答以下问题。

问题:{question}

{format_instructions}"""

prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个专业的数学老师，擅长用简单易懂的方式解释数学概念。"),
    ("user", template)
])

# 创建处理链
chain = prompt | chat | output_parser

# 运行链
rs = chain.invoke({
    'style': '通俗易懂',
    'question': '勾股定理是什么？',
    'format_instructions': output_parser.get_format_instructions()
})

print(rs)
print(f"答案: {rs.answer}")
print(f"置信度: {rs.confidence}")
```

# 流式输出

```python
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.prompts import ChatPromptTemplate
from pydantic import BaseModel, Field

# 定义输出结构（使用 Pydantic）
class MathAnswer(BaseModel):
    answer: str = Field(description="问题的答案")
    confidence: float = Field(description="答案的置信度，范围0-1")

# 创建输出解析器
output_parser = PydanticOutputParser(pydantic_object=MathAnswer)

# 定义提示模板
template = """你是一名数学老师，请用{style}风格回答以下问题。

问题:{question}

{format_instructions}"""

prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个专业的数学老师，擅长用简单易懂的方式解释数学概念。"),
    ("user", template)
])

# 创建处理链
chain = prompt | chat 

# 运行链
for chunk in chain.stream({'style':'通俗易i','question':'勾股定理是什么?','format_instructions': output_parser.get_format_instructions()}):
    print(chunk.content,end='',flush=True)
```
